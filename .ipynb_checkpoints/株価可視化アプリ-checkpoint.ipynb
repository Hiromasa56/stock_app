# yfinanceにより企業のその日の株価の終値を取得する関数

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as   yf           #株価取得のライブラリ

%matplotlib inline
```

```python
def get_data(days, tickers):
    df = pd.DataFrame()  #複数の企業のデータを入れるデータフレーム
    for company in tickers.keys():
    #company = 'facebook'

        tkr = yf.Ticker(tickers[company]) 

        hist = tkr.history(period=f'{days}d') 
        hist.index = hist.index.strftime('%d %B %Y') 
        hist = hist[['Close']]  
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        df = pd.concat([df, hist])     #dfにそれぞれのデータを追加できる。axis=0でくっつける
    return df
```

```python
days = 20

tickers = {
    'apple': 'AAPL',
    'facebook': 'FB',
    'google': 'GOOGL',
    'microsoft': 'MSFT',
    'netflix': 'NFLX',
    'amazon': 'AMZN'
}

df = get_data(days, tickers)
```

## 可視化ライブラリaltair

```python
import altair as alt
```

```python
df
```

```python
companies = ['apple', 'facebook'] #appleとfacebookだけのデータを使う
data = df.loc[companies]
data
```

```python
data.sort_index()   #indexをアルファベット順に並べる
```

```python
data = data.T.reset_index()   #グラフはDateのデータが必要だから、データを成形する
data.head()
```

```python
#melt関数でDateを基準にそれぞれのカラム毎にデータをばらしている
data = pd.melt(data, id_vars=['Date']).rename(
    columns={'value': 'Stock Prices(USD)'}
)                                                
data.head()
```

```python
#可視化の設定(折れ線グラフ)
y_min, y_max=180,230

chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8, clip=True)
    .encode(
        x='Date:T',
        y=alt.Y('Stock Prices(USD):Q', stack=None, scale=alt.Scale(domain=[y_min, y_max])),
        color='Name:N'
    )
)
```
